```python
import pandas as pd
import os
```

```python
datadir = "../data/"
mzMLLoc = "X:/Ethan/integrAitor/raw_data/polar_neg/"
```

```python
metadata = pd.read_csv(datadir + "LLFS_Experiment_parameters_v7.csv",index_col=5)
metadata
```

```python
reformatted_data = {}
for index,row in metadata.iterrows():
    batch = row["Batch analysis"]
    dirBase = datadir + "negative/Batch" + batch + "/"
    fn = dirBase + "peak_areas.csv"
    data = pd.read_csv(fn,index_col=0)
    peakBoundaries = pd.read_csv(dirBase + "rt_bounds_peak-boundaries.csv")
    filt = peakBoundaries[peakBoundaries["File Name"] == index+".d"]
    #os.system("cp " + mzMLLoc + index + ".mzML " + mzMLTarget)
    for met in data.index.values:
        filtfilt = filt[filt["Peptide Modified Sequence"] == met]
        rt_start = filtfilt.at[filtfilt.index.values[0],"Min Start Time"]
        rt_end = filtfilt.at[filtfilt.index.values[0],"Max End Time"]
        peakArea = data.at[met,index + " Area"]
        reformatted_data[len(reformatted_data)] = {"sample":index,"metabolite":met,"rt_start":rt_start,"rt_end":rt_end,"area":peakArea,"path":mzMLLoc + index + ".mzML"}
        if len(filtfilt) != 1:
            print(len(filtfilt))
df = pd.DataFrame.from_dict(reformatted_data,orient="index")
df
```

```python
df.to_csv(datadir + "formatted_data.csv",index=False)
```
